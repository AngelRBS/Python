# Condicional If

# If Else

```python
# Notas de materias
nota = float(input("Ingresa una nota: "))

if nota >= 70: # Si es mayor o igual a 70
  print("Acreditaste la materia!")
else: # Si es menor que 70
  print("Nn acreditaste la materia...")
```

```python
edad = int(input("Ingresa tu edad: "))
if edad >= 18: # Mayor o igual a 18
  print("Eres mayor de edad.")
  if edad >= 65: # Mayor o igual a 65
    print("Ademas, ya eres un adulto mayor")
  else: # Menor a 65
    print("Ademas, eres un adulto menor")
else: # Menor a 18
  print("Eres menor de edad")
  if edad <= 12: # Menor o igual a 12
    print("Ademas, eres un niño")
  else: # Mayor a 12
    print("Ademas, eres un adolescente")
```

# ElIf

Ahora hagamos un programa en el que preguntamos la calificacion del 0 al 100 y queremos saber su equivalente en el sistema de evaluacion estadounidense. Para simplicidad, basemonos en esto:

A+ = 100
A = 90 - 99
B = 80 - 89
C = 70 - 79
D = 60 - 69
F = <60


```python
cal = float(input("Ingresa una calificacion: "))
if cal == 100:
  print("A+")
elif cal >= 90:
  print("A")
elif cal >= 80:
  print("B")
elif cal >= 70:
  print("C")
elif cal >= 60:
  print("D")
else:
  print("E")
```

# Operadores Logicos con Condicionales

```python
# Condicionales: and, or, not
# Quieres un sandwich con todo
pan = True
queso = True
mayonesa = True
jamon = False
print(pan and queso and mayonesa and jamon)
if not (pan and queso and mayonesa and jamon):
  print("No puedes hacer el sandwich.")
```

```python
cal = float(input("Ingresa una calificacion: "))
if cal >= 0 and cal <= 100:
  if cal == 100:
    print("A+")
  elif cal >= 90:
    print("A")
  elif cal >= 80:
    print("B")
  elif cal >= 70:
    print("C")
  elif cal >= 60:
    print("D")
  else:
    print("E")
else:
  print("Error. Calificacion no valida.")
```

```python
# El modulo (%)
# El modulo es un operador matematico que devuelve el residuo de una division
# 10 % 3 da 1, porque 10 dividido entre 3 cabe 3 veces y sobra 1
# 7 % 2 da 1, porque el 7 no es divisible de forma exacta entre 2

num = 9
if num % 3 == 0 or num % 5 == 0:
  print(f"El numero {num} es multiplo de 3 o 5")

# Saber si el numero es par o impar, usando not
num = 6
if not num % 2 == 0:
  print("El numero es impar.")
else:
  print(f"El numero {num} es par.")


```

# Ciclo FOR

Los llamados ciclos for, nos ayudan a realizar acciones durante una cantidad determinada

```python
# Imprimir los numeros del 0 al 4
for i in range(5): # Toma 5 como el numero final
  print(i)
```

```python
for num in range(2,11):
  print(num)
```

```python
# Tambien existe el range(inicio, fin, paso)
for i in range(2,21,2):
  print(i)
```

```python
cadena = "Python Hola"
for letra_por_letra in cadena:
  print(letra_por_letra)
```

# While

```python
suma = 0
numeros = int(input("Ingrese un numero para sumar (Ingrese 0 para salir): "))

while numeros != 0: # != es equivalente a "distinto de... "
  suma += numeros
  numeros = int(input("Ingrese un numero para sumar (Ingrese 0 para salir): "))

print(f"La suma total es: {suma}")

```

```python
# Usar el while como for
numero = 0
while numero <= 12:
  print(numero)
  numero += 1
```
