# ESTRUCTURAS DE DATOS

Las estructuras de datos son formas de organizar información para trabajar con ella de manera eficiente. Veremos las siguientes:

- Listas --> ordenadas, modificables, permiten duplicados.
- Tuplas -->  ordenadas, inmutables, permiten duplicados.
- Sets --> no ordenados, sin duplicados.
- Diccionarios--> pares clave:valor.

# LISTAS

Las listas son colecciones ordenadas que podemos modificar y almacenan elementos de distintos tipos de datos. Se definen usando corchetes [] y los elementos se separan por comas. Veamos un primer ejemplo sobre una lista de compras.

```python
lista_de_compras = ["manzanas", "pan", "leche", "huevos", "mantequilla", "queso","jamón"]
print(lista_de_compras)
```

```python
print(lista_de_compras[6])
```

```python
# Agregar un elemento a la lista .append()
lista_de_compras.append("sopa")
print(lista_de_compras)
```

```python
# Agregar un elemento en una posicion especifica
lista_de_compras.insert(1, "platano")
print(lista_de_compras)
```

```python
print(lista_de_compras)
```

```python
# Para elimiar un elemento de una lista existe el remove y el pop
lista_de_compras.remove("platano")
print(lista_de_compras)
```

```python
lista_de_compras.pop() # Sin nada dentro del parentesis borra el ultimo elemento, en este caso, "sopa"
lista_de_compras.pop(7) # O puedes agarrar la posicion que quieras, siempre iniciando desde 0
# Nota: te retorna el numero eliminado
```

```python
# Podemos recorrer el arreglo por medio de un For
for elementos in lista_de_compras:
  print(elementos)
```

```python
# len saca el tamaño de la lista
for i in range(len(lista_de_compras)):
  print(f"{i+1} - {lista_de_compras[i]}")
```

```python
# Piedra, papel o tijeras
import random
opciones = ['piedra', 'papel', 'tijeras']

while True:
  while True:
    eleccion = input("\nElige piedra, papel, o tijeras (o escribe salir si ya no quieres jugar): ")
    if eleccion not in opciones and not eleccion == 'salir':
      print("Opcion no valida, intenta de nuevo.")
    else:
      break

  if eleccion == 'salir':
    print("Gracias por jugar!")
    break
  aleatoria = random.choice(opciones)
  print("La computadora eligio: ", aleatoria)

  # Determinamos el ganador
  if eleccion == aleatoria:
    print("Empate!")
  elif (eleccion == 'piedra' and aleatoria == 'tijeras') \
       or (eleccion == 'papel' and aleatoria == 'piedra') \
       or (eleccion == 'tijeras' and aleatoria == 'papel'):
    print("Ganaste!")
  else:
    print("Perdiste... :(")





```

```python
promedios = [55.6, 97.5, 23.6, 97.5, 72.4, 100]
print(promedios)

promedios.sort() # Sirve para ordenar la lista de manera ascendente
print(promedios)

promedios.sort(reverse=True) # Sirve para ordenar la lista de manera descendente
print(promedios)
```

```python
# Podemos ingresar el ultimo elemento de una lista
print(promedios[-1])

```

```python
# Encontrar el mayor y el menor de una lista
print(max(promedios))
print(min(promedios))
```

```python
# Declaracion y uso de matrices
calificaciones = [[100, 88, 93], [99, 56, 72]]
i = 1
for cali_i in calificaciones:
  print(f"\nAlumno {i}")
  j = 1
  for cali_j in cali_i:
    print(f" Cali {j}: {cali_j}")
    j =+ 1
  j = 0
  i += 1
```

# SETS
Los sets son colecciones SIN orden y que no permiten elementos duplicados

```python
nombre = {'Angel', 'Eliud', 'Diana', 'Juan', 'Juan'}
print(nombre)
print(type(nombre))
```

```python
distintos_datos = {'Angel', 1, 'Diana', 'Juan', 'Juan', 4.8}
print(distintos_datos)
```

```python
# Add
nombre.add("Leonardo")
print(nombre)
```

```python
print(nombre[2])
```

```python
nombre[2] = 'Axel'
```

```python
# Union
matutino = {20322, 22183, 123321, 392873, 23420}
vespertino = {16432, 577753, 32245, 246543, 20322}
# Union pero con .union()
total_matriculas = matutino.union(vespertino)
print(total_matriculas)
# Union pero con |
total_matriculas = matutino | vespertino
print(total_matriculas)
```

```python
# Interseccion
ca = {1234, 123, 4223, 532, 6453, 689, 109}
al = {1234, 11, 22, 33, 44, 55}
# Interseccion con .intersection()
ca_al = ca.intersection(al)
print(ca_al)
# Interseccion usando &
ca_al = ca & al
print(ca_al)
```

```python
# Diferencia
# Diferencia usando .difference()
solo_calculo = ca.difference(al)
print(solo_calculo)
# Diferencia usando -
solo_calculo = ca - al
print(solo_calculo)
```

```python
nombre = {'Angel', 'Eliud', 'Diana', 'Juan', 'Juan'}
# remove, pop
nombre.remove('Angel') # Puedes eliminar el que tu quieras
print(nombre)
nombre.pop() # Elimina un elemeto de manera aleatoria
print(nombre)
```

```python
# Eliminar todo el SET
nombre.clear()
print(nombre)
```

# Diccionarios
Los diccionarios son colecciones no ordendadas de pares clave:valor

```python
alumno = {
    # clave      valor
    "nombre": "Eliud Sanchez",
    "edad": 21,
    "carrera": "LCC",
    "semestre": 7,
    "promedio": 88,
    "calificaciones": [100, 94, 94, 97, 100]
}

print(type(alumno))
print(alumno)
```

```python
print("El alumno ", alumno["nombre"], "esta estudiando ", alumno["carrera"])
```

```python
# Agregar al diccionario
alumno["area preferida"] = "Ciencia de datos"
print(alumno)
```

```python
for valor in alumno:
  print(alumno[valor])
```

```python
for clave, valor in alumno.items():
  print(clave, ": ", valor)
  # Devuelve una vista de todos los pares clave:valor del diccionario
```

```python
# El metodo .clear() tambien se puede usar con diccionarios
# Tenemos el .get que nos ayuda a obtener el valor (value) de cierta clave (key)
print(alumno.get("nombre"))
```

```python
alumno_1 = alumno.items()
print(alumno_1)
print(type(alumno_1))
alumno_lista = list(alumno_1)
print(alumno_lista)
```

```python
# El metodo .values() sirve para obtener los valores del diccionario, tambien se puede usar en la interaccion
alumno_v = alumno.values()
print(type(alumno_v))
print(alumno_v)
```

```python
if "carrera" in alumno:
  print("Existe carrera")
else:
  print("No existe carrera")

if "dinero" in alumno:
  print("El alumno tiene dinero")
else:
  print("El alumno no tiene dinero")
```

```python
# .copy() puedes copiar todo el diccionario en una nueva variable
# Update --> Sirve para cambiar los valores (valores) (.update())
alumno.update({"semestre": 6})
print(alumno)

```
