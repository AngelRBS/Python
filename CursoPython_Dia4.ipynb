## Funciones

En este curso hemos usado algunas funciones integradas por Python, pero tambien nosotros podemos definir funciones, siendo estas un bloque de codigo reutilizable que podemos mandar a llamar varias veces en el codigo.

```python
def mensaje():
  print("Hola!!!")

mensaje()
```

```python
# Funcion con retorno
def mensaje():
  return "Hola!!!"

print(mensaje())

va = mensaje()
print(va)
```

```python
def saludo(name):
  print(f"Hola, ", name, "!!!")
tuNombre = input("Ingresa tu nombre: ")
saludo(tuNombre)
# Nota: El nombre de la variable del parametro puede o no ser el mismo fuera de la funcion
```

```python
# Puedes mandarle el parametro de esta manera
saludo("grupo")
```

```python
def suma(numero_1, numero_2):
  return numero_1 + numero_2

print(suma(1,5))
```

```python
def bienvenida(nombre = "invitado"):
  print("Hola, ", nombre)

bienvenida()
bienvenida("Victor")
```

```python
def suma(numero_1 = 2, numero_2 = 10): # Podemos ponerle tantos parametros como queramos
  return numero_1 + numero_2
print(suma())
suma = (suma(1000, 102339))
print(suma)
```

```python
# Descuento

def descuento(precio, desc = 0.15):
  precio = precio * (1 - desc)
  print("El nuevo precio es: ", precio)

descuento(750)
descuento(800, .25)
```

```python
def usuario(nombre, edad):
  return (nombre, edad)

persona = usuario("Victor", 21)
print(type(persona))
print(persona)
```

```python
def operaciones(numero):
  doble = numero * 2
  triple = numero * 3
  cuadruple = numero * 4
  quintuple = numero * 5
  return doble, triple, cuadruple, quintuple

resultado = operaciones(2)
print(resultado)
```

## Funciones que llaman a otras funciones

```python
def saludar(nombre):
  return f"Hola, {nombre}!"

def despedirse(nombre):
  return f"Adios, {nombre}!"

def mensaje_completo(nombre):
  saludo = saludar(nombre)
  despedida = despedirse(nombre)
  return f"{saludo}\n{despedida}" # \n equivale a un salto de linea

print(mensaje_completo("Angel"))
```

```python
# Ahora supongamos que tenemos una heladeria

def pedir_sabor():
  sabores = ['vainilla', 'chocolate', 'fresa']
  print("Sabores disponibles: ", ",", sabores)
  # lower() hace minusculas las palabras y strip() elimina espacios
  sabor = input("Elige un sabor: ").strip().lower()

  # Si sabor NO esta en sabores, entra
  while sabor not in sabores:
    sabor = input("Elige un sabor nuevamente: ").strip().lower()
  return sabor

def pedir_cantidad():
  cantidad = int(input("Cuantas bolas de nieve?"))
  while cantidad <= 0:
    cantidad = int(input("Cuantas bolas de nieve? (Ingrese un valor valido)"))
  return cantidad

# Funcion madre
def pedir_helado():
  sabor = pedir_sabor()
  cant = pedir_cantidad()
  print(f"Pedido: {cant} bola(s) de helado de {sabor}")
  return cant, sabor

total_pedidos = int(input("Cuantos clientes atenderas?: "))
for i in range(total_pedidos):
  pedir_helado()


```

## Funciones Lambda

Una funcioin Lambda es una funcion acortada. Para escribir una funcion de esta forma:

nombreFuncion = Lamda parametros: accion

La diferencia entre las funciones Lambda y las funciones normales es que suelen ser un poco mas limitadas

```python
from ast import Lambda
sumar = lambda x,y: x + y
print(sumar(2,3))
```

```python
descuentoLambda = lambda desc, precio: ((precio * (1 - desc)), precio * desc)
nuevoPrecio = descuentoLambda(0.4, 500)
print("El precio a pagar es: ", nuevoPrecio[0], ", el descuento que se hizo fue: ", nuevoPrecio[1])
print(type(nuevoPrecio))
```

```python
# Calculadora --> suma, resa, producto, division y potencia
def suma(a, b):
  return a + b
def resta(a, b):
  return a - b
def producto(a, b):
  return a * b
def division(a, b):
  if b == 0:
    return "Error. No se puede dividir entre 0"
  return a / b
def potencia(a, b):
  return a ** b

def menu():
  print("-------- CALCULADORA --------")
  print("1. Suma")
  print("2. Resta")
  print("3. Producto")
  print("4. Division")
  print("5. Potencia")
  print("6. Salir")

while True:
  menu()
  try:
    op = int(input("Seleccione una opcion: "))
    if op < 1 or op > 6:
      print("Opcion no valida. Intente de nuevo")

  except ValueError:
    print("Valor no valido. Ingrese un valor numerico.")

  if op == 6:
    print("Saliendo...")
    print("Gracias por usar el sistema.")
    break

  try:
    a = float(input("Ingrese el primer numero: "))
    b = float(input("Ingrese el segundo numero: "))
  except ValueError:
    print("Valor no valido. Ingrese un valor numerico.")

  if op == 1:
    print("Resultado: ", suma(a, b))
  elif op == 2:
    print("Resultado: ", resta(a, b))
  elif op == 3:
    print("Resultado: ", producto(a, b))
  elif op == 4:
    print("Resultado: ", division(a, b))
  elif op == 5:
    print("Resultado: ", potencia(a, b))
  else:
    print("Opcion no valida.")


```
