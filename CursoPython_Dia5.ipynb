## Que es un modulo?

Un modulo es una unidad de organizacion del codigo que permite encapsular y dividir el software en partes mas manejables, generalmente representando como un archivo que contiene codigo relacionado, como funciones, variables (constantes) o metodos de clase.

JSON es un formato para guardar y transmitir datos. Es muy util en APIs, archivos de configuracion y almacenamiento de informacion.

(Se parece a un diccionario de Python, pero es solo texto)

Para que sirve?
  - Sirve para convertir objetos de Python --> texto JSON y tambien texto JSON --> objetos de Python

  Esto se llama:
  
    - Serializar: Convertir un objeto de Python en texto JSON

    - Deserializar: Convertir un texto JSON a un objeto usable de Python

```python
# Funciones principales:
# 1) json.dums(obj): Convierte un diccionario u objeto de Python en un string JSON
import json
obj = {
    "nombre": "Ana",
    "edad": 25
    }
print(type(obj))
s = json.dumps(obj, ensure_ascii = False)
print(s)
print(type(s))
```

```python
# json.loads(texto_json): Convierte un string JSON en un diccionario Python
restaurar = json.loads(s)
print(restaurar)
print(type(restaurar))
# Al venir de una pagina/persona/archivo desconocid@ debes de tener cuidado
va = '''
Pueden contener:
- Datos malinformados
- Informacion inesperada
- Campos faltantes
- Datos demasiado grandes
- Variables incorrectas que rompan tu programa
'''

```

```python
va = '''
#UTF-8 es la abreviatura de "8-bit Unicode Transformation Format"
y se trata de un formato de codificacion de caracteres Unicode e ISO 10646 que utiliza simbolos
de longitud variable, permitiendo representar cualquier caracter Unicode con una
eficiencia que lo hace ideal para internet. '''

with open("datos.json", "w", encoding='utf-8') as f:
  json.dump(obj, f, ensure_ascii=False, indent=4)
```

## Numpy

Provee el tipo ndarray, operaciones vectorizadas, brodcasting, utilidades de algebra lineal y funciones matematicas en arrays. Es la base sobre la que se apoyan muchas librerias cientificas en Python.

### Nota IMPORTANTE: El arreglo de Numpy no es igual a la lista de Python, este NO puede tener diferentes tipos de datos en el mismo

```python
# pip install numpy
import numpy as np
A = np.array([[1,2,3],[4,5,6]]) # array (2,3)
v = np.array([10,20,30]) # array (3)
print(A + v)
print("-----------")
prueba = [1,2,3]
prueba2 = [1,2,3]
print(prueba + prueba2)
```

```python
# Mas funciones de Numpy
z = np.zeros((3,3)) # array (3,3)
print(z)
o = np.ones((2,4))
print(o)
r = np.random.rand(3,3)
print(r)
```

```python
# Rangos numericos
nums = np.arange(0,10,2) # Saca los valores entre 0 y 10 de 2 en 2
print(nums)
x = np.linspace(0,1,5) # Saca 5 valores espaciados de manera "equitativa" entre 0 y 1
print(x)
```

```python
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,10])
print(a + b) # Suma
print(a - b) # Resta
print(a * b) # Multiplicacion
print(a / b) # Division
print(a ** b) # Potencia
```

```python
print(a.dot(b)) # Producto entre 2 matrices
c = np.array([[1,2,3],[4,5,6]])
d = np.array([[10,20],[30,40],[50,60]])
print(c.dot(d))
```

```python
data = np.array([1,2,3,4,5])
print(data.min()) # Minimo
print(data.max()) # Maximo
print(data.sum()) # Suma total del arreglo
print(data.mean()) # Promedio
print(data.std()) # Desviacion estandar
```

```python
A = np.array([1,6,3,9,2])
print(A[A > 5])
```
